## Call to Action Replace Missing Value

Numerical Column :
1. Lat null value will replace with 42.325559
2. Long null value will replace with -7.107732e+01 which equals to -0.7107732

Categorical Column :
1. All categorical column with null value will replace with mode value from each column itself

## Replace Null value in Numerical column (Lat and Long)

```python
df = df.fillna(df.median())
```

```python
df.isnull().sum()
```

## Replace Null value in Categorical column with mode of each column

```python
df = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
df.isnull().sum()
```

```python
df.describe(include='object')
```

## Fixing space value in REPORTING_AREA column

```python
#Replace '  ' (space) value with '000'
df['REPORTING_AREA'] = df['REPORTING_AREA'].replace({' ' : '000'}, 
                                                    regex=True)
```

```python
df.describe(include='object')
```

```python
df['TIME_OCCURED'] = pd.to_datetime(df['OCCURRED_ON_DATE']).dt.time
df['DATE_OCCURED'] = pd.to_datetime(df['OCCURRED_ON_DATE']).dt.date
df
```
